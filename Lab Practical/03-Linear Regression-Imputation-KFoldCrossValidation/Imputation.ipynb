### Applying Imputation and Testing models

**Data leakage** occurs when information from outside the training dataset influences the model, leading to overly optimistic performance that doesn't generalize to new data. In machine learning, `.fit()` learns the parameters (like mean, variance, etc.) from data. If `.fit_transform()` is applied to both training and test sets, it lets information from the test set leak into the training process, which skews results.

**Correct approach**:
1. Use `.fit()` or `.fit_transform()` **only on the training set** to capture patterns solely from training data.
2. Use `.transform()` on the test set to apply the same transformations without learning new parameters, preventing test data leakage.

```python
X = df.drop('Yield', axis=1)
y = df['Yield']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
print(X_train.isnull().sum().sum())
print(X_test.isnull().sum().sum())
```

```python
mean_imputer = SimpleImputer(strategy='mean').set_output(transform='pandas')
```

Use This

```python
mean_imputer.fit(X_train)
X_imp_train = mean_imputer.transform(X_train)
X_imp_test = mean_imputer.transform(X_test)
print(X_imp_train.isnull().sum().sum())
print(X_imp_test.isnull().sum().sum())
```

Or This

```python
X_imp_train = mean_imputer.fit_transform(X_train)
X_imp_test = mean_imputer.transform(X_test)
print(X_imp_train.isnull().sum().sum())
print(X_imp_test.isnull().sum().sum())
```

```python
linear_model = LinearRegression()
linear_model.fit(X_imp_train, y_train)

# Predict the missing values
y_pred = linear_model.predict(X_imp_test)

r2_score(y_test, y_pred)
```
