```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline  
from sklearn.preprocessing import OneHotEncoder

from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import StackingClassifier

from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, r2_score, roc_auc_score

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'C:\Users\DAI.STUDENTSDC\Desktop\Machine Learning\Data Sets\Cases\Wisconsin\BreastCancer.csv', index_col=0)
```

```python
X = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

Passthrough -> False

```python
nb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=5) 
dtc = DecisionTreeClassifier(random_state=24)  
svm = SVC(probability=True, random_state=24)
lr = LogisticRegression(random_state=24)
```

```python
stack = StackingClassifier(
    estimators=[
        ('knn', knn),
        ('dtc', dtc),
        ('nb', nb),    
        ('svm', svm), 
    ],
    final_estimator=LogisticRegression(random_state=24)
)
stack.fit(X_train, y_train)
```

```python
y_pred = stack.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
y_pred_proba = stack.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred_proba)
```

Passthrough -> True

```python
nb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=5) 
dtc = DecisionTreeClassifier(random_state=24)  
svm = SVC(probability=True, random_state=24)
lr = LogisticRegression(random_state=24)
```

```python
stack = StackingClassifier(
    estimators=[
        ('knn', knn),
        ('dtc', dtc),
        ('nb', nb),    
        ('svm', svm), 
    ],
    final_estimator=LogisticRegression(random_state=24),
    passthrough=True
)
stack.fit(X_train, y_train)
```

```python
y_pred = stack.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
y_pred_proba = stack.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred_proba)
```
